```python
import warnings
warnings.simplefilter('ignore')

%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
%config InlineBackend.figure_format='svg'

from pylab import rcParams
rcParams['figure.figsize']=8,5

df['Survived'].value_counts().plot(kind='bar', label='Survived')
plt.legend()
plt.title('Распределение выживаемости')
```

```python
corr_matrix=df.drop(['Name','Pclass','Sex','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis=1).corr()
sns.heatmap(corr_matrix)
```

```python
features=list(set(df.columns)-set(['PassengerId','Name','Pclass','Sex','Name','SibSp','Parch','Ticket','Cabin','Embarked','Survived']))
df[features].hist(figsize=(10,5))
```

```python
sns.pairplot(df[features+['Survived']], hue='Survived')
```

```python
fig, axes=plt.subplots(nrows=1, ncols=2, figsize=(12,5))

for idx, feat in enumerate(features):
    sns.boxplot(x='Survived', y=feat, data=df, ax=axes[idx])
    #axes[idx].legend()
    axes[idx].set_xlabel('Survived')
    axes[idx].set_ylabel(feat)
```

```python
_, axes=plt.subplots(nrows=1, ncols=2, figsize=(12,5))
sns.boxplot(x='Survived', y="Age", data=df, ax=axes[0])
sns.violinplot(x='Survived', y="Age", data=df, ax=axes[1])
```

```python
sns.countplot(x="Pclass", hue="Survived", data=df)
```

```python
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

X=df.drop(['Name','Pclass','Sex','Name','SibSp','Parch','Ticket','Cabin','Embarked',"Survived"], axis=1)
#X['Survived']=pd.factorize(X['Survived'])[0]

scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)

%time
tsne=TSNE(random_state=17)
tsne_representation=tsne.fit_transform(X_scaled)

plt.scatter(tsne_representation[:, 0], tsne_representation[:, 1])
```
